```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
d = pd.read_csv("C:\\Users\\sampa\\Downloads\\AutoInsurance-updated.csv")
```

```python
d.columns
```

```python
d.head()
```

```python
d_encoded = pd.get_dummies(d, columns=['Coverage','Gender','Marital Status','Vehicle Size'], drop_first=True)
```

```python
correlation_matrix = d_encoded.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", annot_kws={"size": 10})
plt.show()
```

```python
plt.scatter(d_encoded['Monthly Premium Auto'], d_encoded['Customer Lifetime Value'],color='blue', alpha=0.7, label='Data points')
plt.xlabel('Monthly Premium Auto')
plt.ylabel('Customer Lifetime Value')
```

```python
correlation = d_encoded['Monthly Premium Auto'].corr(d_encoded['Customer Lifetime Value'])
correlation
```

```python
d_encoded['logged CLV'] = np.log(d_encoded['Customer Lifetime Value'])
```

```python
d_encoded['logged MPA'] = np.log(d_encoded['Monthly Premium Auto'])
```

```python
correlation = d_encoded['logged MPA'].corr(d_encoded['logged CLV'])
correlation
```

```python
plt.scatter(d_encoded['logged MPA'], d_encoded['logged CLV'],color='blue', alpha=0.7, label='Data points')
plt.xlabel('logged MPA')
plt.ylabel('logged CLV')
```

```python
correlation_matrix_2 = d_encoded.corr()
plt.figure(figsize=(10, 10))
sns.heatmap(correlation_matrix_2, annot=True, cmap='coolwarm', fmt=".2f", annot_kws={"size": 10})
plt.show()
```
